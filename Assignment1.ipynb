```python
from qiskit import QuantumCircuit
qc = QuantumCircuit(3, 3)

qc.h(0)
qc.barrier()

qc.h(1)
qc.cx(1, 2)
qc.barrier()

qc.cx(0, 1)
qc.h(0)
qc.barrier()

a = qc.measure(0, 0)
qc.measure(1, 1)
qc.barrier()

qc.cx(1, 2)
qc.cz(0, 2)

qc.draw(output='mpl')
```
